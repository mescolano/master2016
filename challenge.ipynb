Campaas de Navidad

```python
chunk_it = pd.read_csv('marketing_pax.csv',sep='^',
                       usecols=['revenue_amount_seg','distance_seg','departure_date_leg','booking_status_code',
                                'quality_index'],
                       chunksize = 5e5)
```

```python
%%time
rev_km_xmas = pd.DataFrame()
for (i,chunk) in enumerate(chunk_it):
    chunk = chunk.dropna()
    chunk = chunk[chunk['booking_status_code'] == 'HK']
    chunk = chunk[chunk['quality_index'] >= 0.75]
    chunk['departure_date_leg'] = pd.to_datetime(chunk['departure_date_leg'])
    chunk['Year'] = chunk['departure_date_leg'].map(lambda x: x.year)
    chunk['Month'] = chunk['departure_date_leg'].map(lambda x: x.month)
    chunk['Day'] = chunk['departure_date_leg'].map(lambda x: x.day)
    chunk = chunk[(chunk['Month']==12) & (chunk['Day']>15) & (chunk['Day']<25)]
    chunk.drop(['booking_status_code','quality_index','Month','Day'],axis=1, inplace=True)
    chunk_result = chunk.groupby('Year')['revenue_amount_seg','distance_seg'].sum()
    rev_km_xmas = rev_km_xmas.append(chunk)
rev_km_xmas=rev_km_xmas.groupby('Year')['revenue_amount_seg','distance_seg'].sum()
rev_km_xmas['revenue_per_km'] = rev_km_xmas['revenue_amount_seg']/rev_km_xmas['distance_seg']
```
